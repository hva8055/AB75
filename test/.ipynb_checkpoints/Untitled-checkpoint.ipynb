```python
from flask import Flask, request, render_template
import pandas as pd
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import os

app = Flask(__name__)
UPLOAD_FOLDER = 'uploads'
os.makedirs(UPLOAD_FOLDER, exist_ok=True)
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER

EMAIL_ADDRESS = "hvpubg8055@gmail.com"
EMAIL_PASSWORD = "wzhv pksg hepf sizt"

def process_attendance(attendance_file, email_file):
    csv = pd.read_csv(attendance_file)
    c = csv.drop(index=0)

    att = pd.DataFrame()
    att['Name'] = c['Name']
    att['Roll No'] = c['Roll No']
    
    for i in range(5, 20, 3):
        s = f"Unnamed: {i}"
        if s in c:
            att[i] = c[s]
    
    c_n = ['Name', 'Roll No']
    cc = csv.columns
    for i in range(3, 20, 3):
        if i < len(cc):
            c_n.append(cc[i])
    att.columns = c_n
    
    for col in att.columns[2:]:
        att[col] = pd.to_numeric(att[col], errors='coerce')
    
    below_75 = att[(att.iloc[:, 2:] < 75).any(axis=1)]
    below_75.columns = ['_'.join(col) if isinstance(col, tuple) else col for col in below_75.columns]
    
    email_data = pd.read_csv(email_file)
    merge = pd.merge(email_data, below_75, on='Roll No', how='inner')
    
    return merge

def send_email(receiver_email, student_name, low_attendance_subjects):
    subject = "Attendance Alert"
    body = f"Dear {student_name},\n\nYour attendance is below 75% in the following subjects:\n\n"
    body += "\n".join(low_attendance_subjects)
    body += "\n\nPlease take necessary action."
    
    msg = MIMEMultipart()
    msg["From"] = EMAIL_ADDRESS
    msg["To"] = receiver_email
    msg["Subject"] = subject
    msg.attach(MIMEText(body, "plain"))
    
    try:
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
        server.sendmail(EMAIL_ADDRESS, receiver_email, msg.as_string())
        server.quit()
        print(f"Email sent successfully to {receiver_email}")
    except Exception as e:
        print(f"Error sending email to {receiver_email}: {e}")

@app.route('/')
def upload_form():
    return render_template('upload.html')

@app.route('/upload', methods=['POST'])
def upload_files():
    if 'attendance' not in request.files or 'email' not in request.files:
        return "Please upload both attendance and email CSV files."
    
    attendance_file = request.files['attendance']
    email_file = request.files['email']
    
    att_path = os.path.join(app.config['UPLOAD_FOLDER'], attendance_file.filename)
    email_path = os.path.join(app.config['UPLOAD_FOLDER'], email_file.filename)
    
    attendance_file.save(att_path)
    email_file.save(email_path)
    
    merged_data = process_attendance(att_path, email_path)
    
    for _, row in merged_data.iterrows():
        low_attendance_subjects = [f"{col}: {row[col]}%" for col in merged_data.columns[3:] if row[col] < 75]
        send_email(row['email'], row['Name'], low_attendance_subjects)
    
    return "Emails sent successfully!"

if __name__ == '__main__':
    app.run(debug=True)
```
